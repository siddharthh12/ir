```python
def pagerank(graph, d=0.85, it=100):
    n = len(graph)
    pr = {k: 1 / n for k in graph}
    for _ in range(it):
        new = {}

        for node in graph:
            new[node] = (1 - d) / n + d * sum(
                pr[i] / len(graph[i])
                for i in graph
                if node in graph[i]
            )
        pr = new
    return pr
# Graph
graph = {
    'A': {'B', 'C'},
    'B': {'C'},
    'C': {'A'},
    'D': {'C'}
}
ranks = pagerank(graph)
print("=== PageRank Scores ===")
for k in ranks:
    print(k, round(ranks[k], 6))
```
