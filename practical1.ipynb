```python
# python -m pip install numpy scikit-learn matplotlib scipy seaborn notebook
from collections import defaultdict

docs = {
    1: "She likes to go to school and the park",
    2: "Aman went to play cricket at his school playground",
    3: "Little children like to play at the playground",
    4: "Can you take care of my little one and he likes to play ?"
}

index = defaultdict(set)
for id in docs:
    for word in docs[id].split():
        index[word].add(id)

def search(q):
    words = q.split()
    result = index[words[0]]
    
    for w in words[1:]:
        result = result & index[w]
    
    return sorted(result)

queries = ["playground", "likes", "little"]

for q in queries:
    r = search(q)
    print(f"Query: {q} -> {r}")
```
